```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/humzakhan/Metis/Metis_Projects/Metis_Project_4/processedDF.csv')
df.head()
```

```python
seasons = df['season'].unique()
```

```python
sns.countplot(y='Topic_LDA', data=df, color='blue')
plt.ylabel('Resolution_Category')
plt.savefig('total_df_topics.png')
plt.show()
```

```python
for season in seasons:
    ross_df = df[(df['char'] == 'Ross') & (df['season'] == season)]
    sns.countplot(y='Topic_LDA', data=ross_df, color='blue')
    plt.ylabel('Resolution_Category')
    plt.savefig('ross'+str(season)+'topics.png')
    plt.show()
```
